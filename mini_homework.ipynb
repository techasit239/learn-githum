```python
fig = px.scatter(df,
                x='sugars',
                y='rating',
                #hover_name='name',
                color = 'shelf',
                size = 'cups',
                hover_data=['calories','carbo'],
                hover_name= df['name'].astype(str)+" : "+df['mfr'].astype(str),
                title='Cereal_ratings VS Sugars')
fig.show()
```

```python
df = pd.read_csv("CausesOfDeath_France_2001-2008.csv")
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
# Remove spaces from the 'Value' column and convert to numeric, coercing errors to NaN and filling with 0
df['Value'] = pd.to_numeric(df['Value'].str.replace(' ', '', regex=False), errors='coerce').fillna(0).astype(int)

fig = px.treemap(df, path=['TIME', 'SEX','ICD10'], values='Value', title='dead')   #title is information
fig.show()
```

```python
sum_by_year = df.groupby("TIME", as_index=False)["Value"].sum()
```

```python
sum_by_year
```

```python
sum_by_year = sum_by_year.sort_values("Value", ascending=False)
print(sum_by_year)
```

```python
fig = px.bar(
    sum_by_year,
    x="TIME",
    y="Value",
    title=" (20012008)",
    labels={"TIME": "", "Value": ""},
    text_auto=True
)
fig.show()
```

```python
sum_by_cause = df.groupby("ICD10", as_index=False)["Value"].sum()
sum_by_cause
```

```python
sum_by_cause = sum_by_cause.sort_values("Value", ascending=False)
print(sum_by_cause.head(11))
```

```python
fig = px.bar(
    sum_by_cause.head(11),
    x="Value",
    y="ICD10",
    orientation="h",
    title="Top 10  ( 20012008)",
    labels={"Value": "", "CAUSES": ""},
    text_auto=True
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```
